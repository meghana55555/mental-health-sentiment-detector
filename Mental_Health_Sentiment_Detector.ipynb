```python
# Mental Health Sentiment Detector using VADER
# Step 1: Install and import required libraries
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# Download VADER lexicon (run once)
nltk.download('vader_lexicon')

# Step 2: Initialize Sentiment Analyzer
sia = SentimentIntensityAnalyzer()

# Step 3: Define function to detect mood based on VADER score
def detect_mood_vader(text):
    score = sia.polarity_scores(text)
    compound = score['compound']
    if compound >= 0.05:
        return 'happy ðŸ˜Š'
    elif compound <= -0.05:
        return 'sad ðŸ˜ž'
    else:
        return 'neutral ðŸ˜'

# Step 4: Test with sample mental health related sentences
samples = [
    "I am feeling great today!",
    "I'm so tired and overwhelmed.",
    "Not good at all, feeling down.",
    "I am excited about my future!",
    "I feel stressed and anxious.",
    "Life is beautiful and hopeful.",
    "I have been crying all day.",
    "Everything is just okay."
]

print("Sample Sentiment Predictions:\n")
for text in samples:
    mood = detect_mood_vader(text)
    print(f"Text: {text}\nPredicted Mood: {mood}\n")

# Step 5: Interactive user input loop
print("Enter your feelings below (type 'exit' to quit):")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'exit':
        print("Take care! Goodbye.")
        break
    mood = detect_mood_vader(user_input)
    print(f"Predicted Mood: {mood}\n")

# Step 6: (Optional) Visualization of sample data sentiments
scores = [sia.polarity_scores(t)['compound'] for t in samples]
moods = [detect_mood_vader(t) for t in samples]

plt.figure(figsize=(8,5))
plt.barh(samples, scores, color=['green' if s>0 else 'red' if s<0 else 'gray' for s in scores])
plt.xlabel('Sentiment Score (compound)')
plt.title('Sample Text Sentiment Scores')
plt.show()
```
